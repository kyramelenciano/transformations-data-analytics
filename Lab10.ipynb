### Task 1 (1 mark)

Read the file __file1.txt__ using both __read_csv__ and __read_table__. There are actually two ways of reading it via __read_csv__. Explain the differences you observed while reading the file via these two methods.

```python
### Write your code below this comment.
import numpy as np
import pandas as pd
file_csv_1 = pd.read_csv('data/file1.txt')
file_csv_2 = pd.read_csv('data/file1.txt', sep = "\t")
file_table = pd.read_table('data/file1.txt')
display(file_csv_1.head())
display(file_csv_2.head())
display(file_table.head())
```

#### Explain the differences in the cell below:

What I observed is when using read_csv without specifying the separator, since the default separator for csv is comma, the whole dataset is read as 1 column, because file1.txt is separated by tabs. When I specified the separator as tab, then the data set was read correctly.

When I used read_table, even without specifying the separator, the file was read correctly, because the default separator for read_table is tabs.

### Task 2 (3 marks)

The __wine__ data has been downloaded from the data repository at the University of California Irvine. Read the data from the file __wine.data__ while reading column names and class information from __wine.names__ programmatically (the class attribute is the first column). You can use either __read_csv__ or __read_table__ to read the data. And you can save the class attribute as a column named __'class'__.

```python
### Write your code below this comment.
import csv
with open('data/wine.names') as f:
    lines = list(csv.reader(f))
headers = [lines[4][0].split(" ")[0]] + lines[1]
wine = pd.read_csv('data/wine.data', names=headers)
wine.head()
```

### Task 3 (2 marks)

Make a dictionary for your data such that the columns come from Task 2 above while data comes from __wine.data__. Each column should be a key in your dictionary while the data for that column from __wine.data__ should appear as a tuple of values.

```python
### Write your code below this comment.
with open('data/wine.data') as f:
    lines = list(csv.reader(f))
lines_tuple = []
for line in lines:
    tpl = tuple(line)
    lines_tuple.append(tpl)
lines_tuple
header,values = headers, lines_tuple
wine_dict = {h: v for h, v in zip(header, zip(*values))}
wine_dict
```

### Task 4 (2 mark)

Convert your data dictionary from Task 3 above to JSON. Then save it as a Dataframe named __data_df__ and print it back as JSON such that the rows in the Dataframe correspond with the JSON output.

```python
### Write your code below this comment.
import json
wine_json = json.dumps(wine_dict)
data_df = pd.read_json(wine_json)
print(data_df.to_json(orient='records'))
```
