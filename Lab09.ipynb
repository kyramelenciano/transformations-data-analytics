### Task 1 (3 marks)

Join the three dataframes using the _outer_ and _inner_ mechanisms via the __join__, __concat__, and __merge__ methods. The output from these three methods should look exactly the same.

#### Outer Join

```python
### Write your code below this comment.
# Outer merge
df_new_left_om = pd.merge(df_left, df_right1, left_index=True, right_index=True,how='outer')
df_final_om = pd.merge(df_new_left_om, df_right2, left_index=True, right_index=True, how='outer')
df_final_om
```

#### Inner Join

```python
### Write your code below this comment.

# Inner merge
df_new_left_im = pd.merge(df_left, df_right1, left_index=True, right_index=True, how='inner')
df_final_im = pd.merge(df_new_left_im, df_right2, left_index=True, right_index=True, how='inner')
df_final_im
```

```python
# Inner concat
pd.concat([df_left,df_right1,df_right2], axis=1,  join='inner')
```

### Task 5 (4 marks)

Merge the Dataframe __df_surveys__ with __df_species__ using the _inner_ as well as _outer_ mechanism. For this, you would need to figure out the columns in both Dataframes on which they could be merged. I know you can visually spot the potential matching columns from both Dataframes, but suppose the number of columns in both is so large that the manual approach isn't feasible. So, I would like you to write a function named __pot_joins__ that can identify the potential join columns from both Dataframes. The function should take two Dataframes and return matching columns from both as a list of tuples.

Once you have identified the potential columns for merging, before proceeding with the merge, verify whether the merge columns for both Dataframes contain only a unique value for each row. Also, while merging the two Dataframes, verify the type of the merge. Store the result as __df_merged__. 

<u>Note:</u> You should be able to do this using __merge__ in two different ways. Figure out what those different ways are.

```python
### Write your code below this comment.

def pot_joins(df,df1):
    joins = list()
    for column in df:
        for column2 in df1:
            if column2 in column:
                joins.append((column,column2))
    return joins
            
pot_joins(df_surveys,df_species)
# From the columns identified, we can use species_id (from df_surveys) and id (from df_species) to merge the dfs
```
